```python
import torch
from transformers import AutoTokenizer, AutoModel
```

```python
model = "sentence-transformers/all-MiniLM-L6-v2"
tokenizer = AutoTokenizer.from_pretrained(model)
model = AutoModel.from_pretrained(model)
```

```python
input = "the name is dog. the first of its name."
encoded_inp = tokenizer([input], padding=True, truncation=True, return_tensors="pt")
encoded_inp
```

```python
with torch.no_grad():
  model_outp = model(**encoded_inp)
```

```python
token_embeddings = model_outp.last_hidden_state
print("Token embeddings dim: ", token_embeddings.size())
```
